```python
import matplotlib.pyplot as plt

print(plt.style.available)
```

```python
import numpy as np

plt.style.use('fast')

x = np.arange(10)
for i in range(1, 9):
    plt.plot(x, i * x**3, label='Group %d' % i)
plt.legend(loc='best')
plt.show()
```

```python
with plt.style.context('grayscale'):
    for i in range(1, 9):
        plt.plot(x, i * x**3, label='Group %d' % i)
    plt.legend(loc='best')
    plt.show()
```

```python
import math

n = len(plt.style.available)
num_rows = math.ceil(n/9)

fig = plt.figure(figsize=(20, 20))

for i, s in enumerate(plt.style.available):
    with plt.style.context(s):
        ax = fig.add_subplot(num_rows, 9, i+2)
        for i in range(1, 9):
            ax.plot(x, i * x**3, label='Group %d' % i)
            ax.set_xlabel(s, color='blue')
            ax.legend(loc='best')

fig.tight_layout()
plt.show()
```
